## Advent of Code
Day 6: Trash Compactor

```python
file = open("input.txt", "r")
contents = file.read()

line_list = []
for line in contents.split("\n"):
    if line[0] in ["*", "+"]:
        operators = [each_operator for each_operator in line.split()]
    else:
        line_list.append(line)

totals = []
number_lines = []
for each_operator in operators:
    if each_operator == "*":
        totals.append(1)
    else:
        totals.append(0)
```

```python
# Part 1.

for line in contents.split("\n"):
    if line[0] not in ["*", "+"]:
        for pos, term_str in enumerate(line.split()):
            term = int(term_str)
            if operators[pos] == "*":
                totals[pos] = totals[pos] * term
            else:
                totals[pos] = totals[pos] + term

print(sum(totals))
```

```python
# Part 2.

def pad_ends(line_list) -> list:
    longest_string = len(max(line_list, key=len))

    padded_lines = []
    for each_line in line_list:
        padded_lines.append(each_line.ljust(longest_string))
    return padded_lines


def rotate(line_list) -> list:
    padded_lines = pad_ends(line_list)

    new_str = ""
    for y in range(len(padded_lines[0]) -1, -1, -1):
        for x in range(len(padded_lines)):
            new_str += padded_lines[x][y]
        new_str += "\n"

    return new_str


def has_digit_any(line):
    return any(char.isdigit() for char in line)


def group_lines(new_str: str) -> list:
    output = []
    this_list = []

    for line in new_str.split("\n"):
        if has_digit_any(line):
            this_list.append(int(line))
        else:
            output.append(this_list)
            this_list = []
    return output


backwards_operators = operators.copy()
backwards_operators.reverse()
new_str = rotate(line_list)

new_lines = group_lines(new_str)

grand_total = 0

for term_list in new_lines:
    operator = backwards_operators.pop(0)

    if operator == "*":
        this_total = 1
    else:
        this_total = 0

    for term in term_list:
        if operator == "*":
            this_total = this_total * term
        else:
            this_total = this_total + term
    grand_total += this_total

print(grand_total)
```
