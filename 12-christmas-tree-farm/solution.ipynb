## Advent of Code

Day 12: Christmas Tree Farm.

```python
from icecream import ic

file = open("input.txt", "r")
contents = file.read()

total = 0
for line in contents.split("\n"):
    if "x" in line:
        size_str, quant_str = line.split(": ")

        x_str, y_str = size_str.split("x")
        available = int(x_str) * int(y_str)

        presents = [int(count) for count in quant_str.split(" ")]

        sharing = min(presents[4], presents[5])
        non_sharing = max(presents[4], presents[5]) - min(presents[4], presents[5])

        others = presents[0] + presents[1] + presents[2] + presents[3] + non_sharing

        occupied = 9 * others + 16 * sharing

        fit = occupied <= available

        if fit:
            total += 1

ic(total)
```
