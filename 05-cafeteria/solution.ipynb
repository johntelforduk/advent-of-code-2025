# Advent of Code
Day 5: Cafeteria

```python
# Load data into ranges list of upper and lower tuples.

file = open("input.txt", "r")
contents = file.read()

ranges_str, ingredients_str = contents.split("\n\n")
ranges = []
for each_str in ranges_str.split("\n"):
    lower_str, upper_str = each_str.split("-")
    lower, upper = int(lower_str), int(upper_str)
    ranges.append((lower, upper))
```

```python
# Part 1. Count the ingredients in range.

count = 0
for each_str in ingredients_str.split("\n"):
    ingredient = int(each_str)

    in_range = False
    for lower, upper in ranges:
        if lower <= ingredient <= upper:
            in_range = True
    if in_range:
        count += 1

print(count)
```

```python
def combine_two_ranges(l1, u1, l2, u2) -> tuple|None:
    if l1 <= l2 and u1 >= u2:
        return l1, u1

    elif l2 <= l1 and u2 >= u1:
        return l2, u2

    elif l1 <= l2 <= u1 and u1 <= u2:
        return l1, u2

    elif l2 <= l1 <= u2 and u2 <= u1:
        return l2, u1

    return None

assert combine_two_ranges(1, 2, 5, 6) is None
assert combine_two_ranges(5, 6, 1, 2) is None

assert combine_two_ranges(1,10, 3, 6) == (1, 10)
assert combine_two_ranges(3, 6, 1, 10) == (1, 10)

assert combine_two_ranges(1, 5, 3, 6) == (1, 6)
assert combine_two_ranges(3, 6, 1, 5) == (1, 6)
```

```python
def remove_one_overlap_pair(ranges: list) -> list:
    for pos1, t1 in enumerate(ranges):
        for pos2, t2 in enumerate(ranges):
            if pos1 != pos2:
                l1, u1 = t1
                l2, u2 = t2

                combined = combine_two_ranges(l1, u1, l2, u2)
                if combined is not None:
                    new_ranges = ranges.copy()
                    new_ranges.remove((l1, u1))
                    new_ranges.remove((l2, u2))
                    new_ranges.append(combined)
                    return new_ranges
    return ranges

assert remove_one_overlap_pair([(1, 2), (2, 3), (3, 4)]) == [(3, 4), (1, 3)]
assert remove_one_overlap_pair([(3, 4), (1, 3)]) == [(1, 4)]
assert remove_one_overlap_pair([(1, 2), (20, 30), (40, 50)]) == [(1, 2), (20, 30), (40, 50)]
```

```python
# Part 2.

prev_len = 0
while prev_len != len(ranges):
    prev_len = len(ranges)
    ranges = remove_one_overlap_pair(ranges)

total = 0
for l1, u1 in ranges:
    print(".", end="")
    total += u1 - l1 + 1

print()
print(total)
```
