## Advent of Code
Day 7: Laboratories

```python
file = open("test.txt", "r")
contents = file.read()


def render(g: dict):
    curr_y = 0
    for x, y in g:
        if y != curr_y:
            print()
        print(g[(x, y)], end="")
        curr_y = y
    print("\n")


# Load file contents into a grid dictionary.
grid = {}
y = 0
for line in contents.split("\n"):
    x = 0
    for cell in line:
        grid[(x, y)] = cell
        x += 1

    y += 1

beams = set()
x = 0
while len(beams) == 0:
    if grid[(x, 0)] == "S":
        beams.add(x)
    x +=1

print(beams)

lit_grid = grid.copy()

y = 1
x = 0
splits = 0

while (0, y) in lit_grid:
    while (x, y) in lit_grid:
        # print(grid[(x, y)], grid[(x, y - 1)] )
        if grid[(x, y)] == "^" and x in beams:
            beams.remove(x)
            beams.add(x - 1)
            beams.add(x + 1)
            splits += 1

            lit_grid[(x - 1, y)] = "|"
            lit_grid[(x + 1, y)] = "|"

        elif x in beams:
            lit_grid[(x, y)] = "|"

        x += 1
    x = 0
    y += 1

render(lit_grid)
print(splits)
```
